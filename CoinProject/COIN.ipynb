```python
import pandas as pd
import numpy as np
import scipy.stats as ss

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

np.random.seed(21)
```

```python
# import choices_converted.csv as a pandas dataframe
choices_df = pd.read_csv('choices_converted.csv')
choices_df = choices_df.iloc[:, 1:]

choices_df.head()
```

```python
# get the first 5 columns of choices_df
baseline_choices_df = choices_df.iloc[:, :5]
baseline_choices_df.head()
```

```python
# put all the columns into one column
baseline_choices_df = pd.melt(baseline_choices_df, value_vars=baseline_choices_df.columns)
baseline_choices_df
```

```python
# add 3 columns between variable and value called 'c1' 'c2' 'c3'
baseline_choices_df['c1'] = " "
baseline_choices_df['c2'] = " "
baseline_choices_df['c3'] = " "
baseline_choices_df
```

```python
# set all values for c1, c2, and c3 for variable 'r1_choice' to LL MM HH respectively
baseline_choices_df.loc[baseline_choices_df['variable'] == 'r1_choice', 'c1'] = 'LL'
baseline_choices_df.loc[baseline_choices_df['variable'] == 'r1_choice', 'c2'] = 'MM'
baseline_choices_df.loc[baseline_choices_df['variable'] == 'r1_choice', 'c3'] = 'HH'

baseline_choices_df.loc[baseline_choices_df['variable'] == 'r2_choice', 'c1'] = 'HL'
baseline_choices_df.loc[baseline_choices_df['variable'] == 'r2_choice', 'c2'] = 'LM'
baseline_choices_df.loc[baseline_choices_df['variable'] == 'r2_choice', 'c3'] = 'MH'

baseline_choices_df.loc[baseline_choices_df['variable'] == 'r3_choice', 'c1'] = 'HL'
baseline_choices_df.loc[baseline_choices_df['variable'] == 'r3_choice', 'c2'] = 'LH'
baseline_choices_df.loc[baseline_choices_df['variable'] == 'r3_choice', 'c3'] = 'MM'

baseline_choices_df.loc[baseline_choices_df['variable'] == 'r4_choice', 'c1'] = 'HL'
baseline_choices_df.loc[baseline_choices_df['variable'] == 'r4_choice', 'c2'] = 'LH'
baseline_choices_df.loc[baseline_choices_df['variable'] == 'r4_choice', 'c3'] = 'MM'

baseline_choices_df.loc[baseline_choices_df['variable'] == 'r5_choice', 'c1'] = 'MH'
baseline_choices_df.loc[baseline_choices_df['variable'] == 'r5_choice', 'c2'] = 'LM'
baseline_choices_df.loc[baseline_choices_df['variable'] == 'r5_choice', 'c3'] = 'HL'

baseline_choices_df
```

```python
X = baseline_choices_df[['c1', 'c2', 'c3']]
y = baseline_choices_df['value']
```

```python
# one hot encode the X dataframe
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
label_encoder.fit(y)
```

```python
label_encoder.transform(y)
```

```python
print("Original labels:", label_encoder.classes_, "\n")

print("Mapping from nominal to numerical labels:")
print(dict(zip(label_encoder.classes_,label_encoder.transform(label_encoder.classes_))))
```

```python
baseline_choices_df['value'] = label_encoder.transform(y)
```

```python
baseline_choices_df.tail()
```

```python
y = baseline_choices_df['value']
X = baseline_choices_df[['c1', 'c2', 'c3']]
```

```python
print("X shape: ", X.shape)
print("y shape: ", y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=21)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
baseline_nb = GaussianNB()
baseline_nb.fit(X_train, y_train)
```

```python
X_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
predictions_train = dtc.predict(X_train)
```

```python
predictions_train
```

```python
def compute_accuracy(predictions, actual):
    return (np.sum(predictions == actual) / len(actual)) * 100
```

```python
print("Training accuracy: ", compute_accuracy(y_train, predictions_train),"%")
```

```python
predictions = dtc.predict(X_test)
```

```python
print("Testing accuracy: ", compute_accuracy(y_test, predictions),"%")
```
